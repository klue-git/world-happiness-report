### Check for Missing values

```python
# Group dataset by year and count countries
d1_countrycount = d1_full.groupby('year')['country name'].count()
```

```python
# countries participating per year
d1_countrycount
```

```python
d1_countrycount.to_clipboard()
```

```python
# Group dataset by country and count years
d1_yearcount = d1_full.groupby('country name').agg({'year':['count']})
```

```python
d1_yearcount_sorted = d1_yearcount.sort_values(('year', 'count'),ascending=False)
```

```python
# Years countries participated in
d1_yearcount_sorted
```

```python
d1_yearcount_sorted.to_clipboard()
```

```python
# Limit D1 to countries that have observations in 2017 - 2020
# Create Crosstab for countries and years
crosstab = pd.crosstab(d1_full['country name'], d1_full['year'], dropna=False)
```

```python
crosstab.to_clipboard()
```

```python
# Filter out the years 2017-2022
filtered_years = crosstab.loc[:, 2017:2022]
```

```python
# Find countries with 0 entries for those years
countries_with_zero_entries = filtered_years[filtered_years.prod(axis=1) == 0].index.tolist()
```

```python
countries_with_zero_entries
```

```python
count_zero_entries = sum(1 for country in countries_with_zero_entries)
```

```python
count_zero_entries
```

```python
# Exclude those countries from the original dataframe d1_full
d1_filtered = d1_full[~d1_full['country name'].isin(countries_with_zero_entries)]
```

```python
# drop entries for the years before 2017 from filtered list
d1_filtered = d1_filtered[d1_filtered['year'] >= 2017]
```

```python
d1_filtered.describe()
```

#### Crosstab check shows 89 eligible countries to compare within the time frame of 2017 - 2022. 76 countries will be excluded from further analysis. As will years before 2017.

```python
# Append 2023 and 2020 d2 datasets and create crosstab to identify missing countries
d2_comp = pd.concat([d2_2023, d2_2020])
```

```python
d2_comp
```

```python
# Create Crosstab
crosstab_d2 = pd.crosstab(d2_comp['country'], d2_comp['year'], dropna=False)
```

```python
crosstab_d2
```

```python
# Find countries with 0 entries for those years
countries_with_zero_entries_d2 = crosstab_d2[crosstab_d2.prod(axis=1) == 0].index.tolist()
```

```python
countries_with_zero_entries_d2
```

```python
# Exclude those countries from the d2_comp dataframe
d2_comp_filtered = d2_comp[~d2_comp['country'].isin(countries_with_zero_entries_d2)]
```

```python
d2_comp_filtered['year'].value_counts()
```

```python
d2_comp_filtered.shape
```

```python
# Create filtered single version for d2_2023
d2_2023_filtered = d2_2023[~d2_2023['country'].isin(countries_with_zero_entries_d2)]
```

```python
# Create filtered single version for d2_2020
d2_2020_filtered = d2_2020[~d2_2020['country'].isin(countries_with_zero_entries_d2)]
```

```python
# Concatenate D2 datasets for future visualisations
d2_full = pd.concat([d2_2023,d2_2022,d2_2021,d2_2020,d2_2019,d2_2018,d2_2017,d2_2016,d2_2015])
```
